```python
from pypdf import PdfReader
from tqdm.auto import tqdm

def get_pdf_text(filename, password='password'):
    pdf_text = ''
    # Open the PDF file in binary mode
    with open(filename, 'rb') as pdf_file:
        # Create a PDF reader object
        pdf_reader = PdfReader(pdf_file)
        # Check if the PDF is encrypted
        if pdf_reader.is_encrypted:
            # If it's encrypted, you may need to provide a password to decrypt it
            pdf_reader.decrypt(password)
        # Get the number of pages in the PDF
        num_pages = len(pdf_reader.pages)
        # Loop through each page and extract text
        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            text = page.extract_text()
            # You can process the text as needed here
            pdf_text += text
    # Remember to close the PDF file
    pdf_file.close()
    return pdf_text

tqdm.pandas(desc='Get PDF text')
df['text'] = df['filename'].progress_apply(get_pdf_text)
df['text'].head()
```

```python
def remove_page_numbers(text):
    # Define a regular expression pattern to match "pag. N" where N is one or more digits
    pattern = r'pag\.\s*\d+'
    # Use re.sub to replace all matches with an empty string
    cleaned_text = re.sub(pattern, '', text)
    # Define a regular expression pattern to match "Pagina N di M" where N is one or more digits
    pattern = r'Pagina\s*\d+\s*di\s*\d+'
    # Use re.sub to replace all matches with an empty string
    cleaned_text = re.sub(pattern, '', cleaned_text)
    return cleaned_text

def clean_text(text, remove_newlines=False):
    text = text.replace('SEGRETARIATO © Copyright Giuffrè Francis Lefebvre S.p.A.', '')
    text = text.replace('© Copyright Wolters Kluwer Italia s.r.l.', '')
    text = text.replace('WOLTERS KLUWERONE LEGALE\n\n', '')
    text = text.replace('One LEGALE', '')
    text = text.replace('Ottobre 2023', '')
    # Remove consecutive spaces
    # text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[ ]+', ' ', text)
    text = text.replace('\t', ' ')
    text = text.replace('\' ', '\'')
    # Remove consecutive newlines
    text = re.sub(r'\n+', '\n', text)
    text = text.replace('\n\n', ' ')
    # text = re.sub(r'\n{2,}', '\n', text)
    text = remove_page_numbers(text)
    if remove_newlines:
        text = text.replace('\n', ' ')
    return text.strip()

tqdm.pandas(desc='Clean text')
df['text'] = df['text'].progress_apply(clean_text)
df['text'].head()
```

```python
def get_sections(
    text,
):
    sections = [
        'Svolgimento del processo',
        'Motivi della decisione',
        'P.Q.M.',
        'Conclusione',
    ]
    incipit = text.split(sections[0])[0].strip()
    if not all([s in text for s in sections]):
        sections = [
            'SENTENZA',
            'RITENUTO IN FATTO',
            'CONSIDERATO IN DIRITTO',
            'P.Q.M.',
        ]
        # Isolate the incipit, which is between 'SENTENZA' and 'SENTENZA'
        incipit = text.split(sections[0])[1].split(sections[1])[0].strip()
        text = text.split(incipit)[-1]
    ret = {}
    if incipit != '':
        ret['Incipit'] = incipit
    for i in range(len(sections) - 1):
        if sections[i] in text and len(text.split(sections[i])) == 2:
            tmp = text.split(sections[i])[-1]
            ret[sections[i]] = tmp.split(sections[i+1])[0].strip()
    ret[sections[-1]] = text.split(sections[-1])[-1].strip()
    return ret

text = df['text'].iloc[0]
print(text[:1000])
print("-" * 80)
for k, v in get_sections(text).items():
    print(f"--> {k}:\n{v[:50]}...\n\n")
```

```python
def concatenate_short_sentences(sentences, length_threshold=10):
    """
    Concatenate short sentences with the previous one if their length is below the specified threshold.

    :param sentences: List of sentences.
    :param length_threshold: Length threshold for a sentence to be considered short.
    :return: Updated list of sentences with short sentences concatenated to the previous ones.
    """
    updated_sentences = []
    for i, sentence in enumerate(sentences):
        if len(sentence) < length_threshold and updated_sentences:
            # Append the current short sentence to the previous one
            updated_sentences[-1] += " " + sentence
        else:
            # Add the current sentence as a new entry in the list
            updated_sentences.append(sentence)

    return updated_sentences

# Adjusting the function to correctly handle sentences split by "n." and re-joining sentences from numbered bullet points
def split_into_sentences(text, length_threshold=20, acronyms=acronyms):
    # Adjusted regular expression to handle more complex splitting scenarios
    potential_sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s', text)

    sentences = []
    current_sentence = ""
    for sentence in potential_sentences:
        # Trim whitespace and check if the sentence is not just empty space
        trimmed_sentence = sentence.strip()
        if trimmed_sentence:
            # Rejoin sentences that were incorrectly split due to specific patterns
            pattern = '|'.join(acronyms)
            if re.search(rf'(\b(?:{pattern})\.$)', trimmed_sentence.lower()) or re.search(r'\d\.$', trimmed_sentence) or re.search(r'^\d+\.\s', trimmed_sentence):
                current_sentence += trimmed_sentence + " "
            else:
                if current_sentence:
                    # # If the current sentence is short, append the next sentence to it
                    # if len(current_sentence) < length_threshold:
                    #     current_sentence += " " + trimmed_sentence
                    # else:
                    #     sentences.append(current_sentence)
                    #     current_sentence = trimmed_sentence

                    current_sentence += trimmed_sentence
                    sentences.append(current_sentence.strip())
                    current_sentence = ""
                else:
                    sentences.append(trimmed_sentence)

    # Adding the last sentence if there's any
    if current_sentence:
        sentences.append(current_sentence.strip())

    sentences = concatenate_short_sentences(sentences, length_threshold)
    sentences = [s.replace('\n', '') for s in sentences if s and len(s) > 5]
    return sentences

# Testing the final improved function on the same text
final_sentences = split_into_sentences(new_text)
# Printing each sentence exactly as split
for idx, sentence in enumerate(final_sentences, start=1):
    print(f"Sentence {idx}: {sentence}")
    if ";" in sentence:
        for j, s in enumerate(list(sentence.split(";")), start=1):
            print(f"\t* Sentence {idx}.{j}: {s}")
    print("")
```

## Rewrite the Text

### Text Utilities

```python
sentence_div = '|||'
section_div = '---'
section_name_div = ':::'

def clean_text(text):
    text = text.replace(sentence_div, "").replace(section_div, "")
    return text.replace(section_name_div, "").strip()

def clean_section_text(text):
    return text.replace(sentence_div, "").strip()

def get_senteces_from_text(text, clean=True):
    sentences = text.split(sentence_div)
    sentences = [s.split(section_name_div)[-1].strip() for s in sentences]
    if clean:
        sentences = [clean_section_text(s) for s in sentences]
    return sentences

def get_sections_from_text(text):
    sections = text.split(section_div)
    return {s.split(section_name_div)[0].strip(): s.split(section_name_div)[-1].strip() for s in sections}

for i, sample in df.iterrows():
    text = sample['text']
    sections = get_sections_from_text(text)
    for k, v in sections.items():
        print(k, clean_section_text(v))
    break
```

```python
import random

random.seed(42)

def chunk_section(text, chunk_size=1024):
    """ Group the sections sentence into chunks of size chunk_size """
    # Split the text into sentences
    sentences = get_senteces_from_text(text)
    # Group the sentences into chunks of size chunk_size
    chunks = []
    curr_chunk = ""
    for sentence in sentences:
        if len(curr_chunk) + len(sentence) < chunk_size:
            curr_chunk += sentence + " "
        elif len(sentence) > chunk_size:
            if curr_chunk != "":
                # print(len(curr_chunk))
                chunks.append(curr_chunk)
            # Keep appending chunks of size chunk_size to the list, if the sentence is longer multiple times than chunk_size
            for i in range(0, len(sentence), chunk_size):
                # print(len(sentence[i:i+chunk_size]))
                chunks.append(sentence[i:i+chunk_size])
            curr_chunk = ""
        else:
            # print(len(curr_chunk))
            chunks.append(curr_chunk)
            curr_chunk = sentence + " "
    # print(len(curr_chunk))
    chunks.append(curr_chunk)
    return chunks

def get_facts(text):
    facts = get_sections_from_text(text)
    if 'Svolgimento del processo' in facts and 'Motivi della decisione' in facts:
        # text = facts['Svolgimento del processo'] + " " + facts['Motivi della decisione']
        text = facts['Svolgimento del processo']
    elif "RITENUTO IN FATTO" in facts:
        text = facts['RITENUTO IN FATTO']
    return text

def get_random_chunk(text, chunk_size=1024):
    """ Get a random chunk of size chunk_size from the text """
    text = clean_section_text(text)
    start = random.randint(0, max(0, len(text) - chunk_size - 1))
    # max_chunks = len(text) // chunk_size
    # start = random.randint(0, max_chunks - 1) * chunk_size
    end = start + min(chunk_size, len(text))
    return text[start:end]

for i, sample in df.iterrows():
    text = sample['text']
    chunk = get_random_chunk(get_facts(text))
    print(chunk)
    break
```

```python
def get_formatted_text(text):
    final_text = ''
    sections = get_sections(text)
    for section, section_text in sections.items():
        sentences = split_into_sentences(section_text)
        final_text += f' --- {section} ::: ' + ' ||| '.join(sentences)
    return final_text

tqdm.pandas(desc='Get formatted text')
df['raw text'] = df['text'].copy()
df['text'] = df['raw text'].progress_apply(get_formatted_text)
```

```python
import pandas as pd

def balance_train_df(train_df, label_columns):
    """ Balance the train_df by oversampling the minority class for each label in label_columns """
    balanced_dfs = []

    for label in label_columns:
        # Get the number of samples with label set to 1 and 0
        label_1_count = train_df[train_df[label] == 1].shape[0]
        label_0_count = train_df[train_df[label] == 0].shape[0]

        # Skip oversampling if all labels are the same
        if label_1_count == 0 or label_0_count == 0:
            balanced_dfs.append(train_df)
            continue

        # Determine the minority and majority class
        minority_class = 1 if label_1_count < label_0_count else 0
        majority_class = 1 if minority_class == 0 else 0

        # Calculate the oversampling factor
        oversampling_factor = label_0_count // label_1_count if minority_class == 1 else label_1_count // label_0_count

        # Oversample the minority class
        minority_df = train_df[train_df[label] == minority_class]
        oversampled_minority_df = pd.concat([minority_df] * oversampling_factor)

        # Combine the oversampled minority class with the majority class
        balanced_df = pd.concat([train_df[train_df[label] == majority_class], oversampled_minority_df])

        # # Print the number of samples with label set to 1 and 0
        # print(f"{label}: {len(balanced_df)} (0/1: {balanced_df[balanced_df[label] == 0].shape[0]} / {balanced_df[balanced_df[label] == 1].shape[0]})")

        balanced_dfs.append(balanced_df)

    # Merge the balanced dataframes for each label
    balanced_train_df = pd.concat(balanced_dfs)

    return balanced_train_df

balanced_df = balance_train_df(train_val_df, label_columns)
tmp = train_val_df[label_columns].sum() - balanced_df[label_columns].sum()

print(len(train_val_df))
print(len(balanced_df))

# Plot tmp with the values on top of the bars
tmp = balanced_df[label_columns].sum()
plt.figure(figsize=(20, 10))
plt.bar(tmp.index, tmp.values)
# Plot the values on top of the bars
for i, v in enumerate(tmp.values):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.xticks(rotation=90)
plt.show()

```

```python
import random

random.seed(42)

def get_chunks(text, chunk_size=1024, min_sentences=3):
    """ Group the sentences into chunks of size chunk_size """
    facts = get_facts(text)
    sentences = get_senteces_from_text(facts)
    # Start from a random sentence and add sentences until the chunk size is reached
    start = random.randint(0, len(sentences) - min(len(sentences), min_sentences))
    chunk = ""
    for i in range(start, len(sentences)):
        if len(chunk) + len(sentences[i]) < chunk_size:
            chunk += sentences[i] + " "
        else:
            chunk += sentences[i][:chunk_size - len(chunk)]
            break
    return chunk.strip()

balanced_df['text'] = balanced_df['text'].apply(get_chunks)
balanced_df
```

```python
def k_fold_tokenized_datasets(
        train_val_df,
        test_df,
        n_splits=5,
        tokenizer=None,
        label_columns=label_columns,
        build_dataset_from_pandas=None,
        preprocess_function=None,
):
    gkf = GroupKFold(n_splits=n_splits)
    k_splits = gkf.split(train_val_df, groups=train_val_df['numero sentenza'])

    test_df_copy = test_df.copy()
    test_df_copy['text'] = test_df_copy['text'].apply(get_chunks)

    if tokenizer is not None:
        pass

    for train_idx, val_idx in k_splits:
        train_df = train_val_df.iloc[train_idx, :].copy()
        # Balance train dataset
        train_df = balance_train_df(train_df, label_columns)
        # Get random chunks from the facts in the train_df text
        train_df['text'] = train_df['text'].apply(get_chunks)
        # Get random chunks from the facts in the val_df text
        val_df = train_val_df.iloc[val_idx, :].copy()
        val_df['text'] = val_df['text'].apply(get_chunks)
        # TODO: Tokenize the two datasets
        if tokenizer is not None:
            pass
        # Yield the train, val and test datasets
        yield train_df, val_df, test_df_copy

for train_df, val_df, test_df in k_fold_tokenized_datasets(train_val_df, test_df, n_splits=5):
    print(f"Train: {len(train_df)} - Val: {len(val_df)} - Test: {len(test_df)}")
```

```python
import random

def get_summary(text):
    summaries = pipe(text)
    return [s['summary_text'] for s in summaries]

def chunk_section(text, chunk_size=1024):
    """ Group the sections sentence into chunks of size chunk_size """
    # Split the text into sentences
    sentences = get_senteces_from_text(text)
    # Group the sentences into chunks of size chunk_size
    chunks = []
    curr_chunk = ""
    for sentence in sentences:
        if len(curr_chunk) + len(sentence) < chunk_size:
            curr_chunk += sentence + " "
        elif len(sentence) > chunk_size:
            if curr_chunk != "":
                # print(len(curr_chunk))
                chunks.append(curr_chunk)
            # Keep appending chunks of size chunk_size to the list, if the sentence is longer multiple times than chunk_size
            for i in range(0, len(sentence), chunk_size):
                # print(len(sentence[i:i+chunk_size]))
                chunks.append(sentence[i:i+chunk_size])
            curr_chunk = ""
        else:
            # print(len(curr_chunk))
            chunks.append(curr_chunk)
            curr_chunk = sentence + " "
    # print(len(curr_chunk))
    chunks.append(curr_chunk)
    return chunks

def get_random_chunk(text, chunk_size=1024):
    """ Get a random chunk of size chunk_size from the text """
    text = clean_section_text(text)
    start = random.randint(0, max(0, len(text) - chunk_size - 1))
    # max_chunks = len(text) // chunk_size
    # start = random.randint(0, max_chunks - 1) * chunk_size
    end = start + min(chunk_size, len(text))
    return text[start:end]


for i, sample in enumerate(dataset):
    facts = get_sections_from_text(sample['text'])
    if 'Svolgimento del processo' in facts and 'Motivi della decisione' in facts:
        # text = facts['Svolgimento del processo'] + " " + facts['Motivi della decisione']
        text = facts['Svolgimento del processo']
    elif "RITENUTO IN FATTO" in facts:
        text = facts['RITENUTO IN FATTO']

    chunk_size = 2048
    chunks = chunk_section(text, chunk_size=chunk_size)
    rand_text = get_random_chunk(text, chunk_size=chunk_size)
    texts = [text, rand_text] + chunks
    print(f"RANDOM: {rand_text[:100]}... (len: {len(rand_text)})")
    print("num chunks:", len(chunks))
    print("max len:", max([len(s) for s in chunks]))

    summaries = get_summary(texts)
    for s, t in zip(summaries, texts):
        print('-' * 80)
        print("ORIGINAL:", t)
        print("SUMMARY: ", s)
    
    if i > 3:
        break
```
